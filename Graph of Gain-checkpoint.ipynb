```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##RESISTOR

freq = [1200, 1000, 600, 300, 200, 150, 100]
voltIn = [0.67423, 0.67537, 0.67983, 0.69067, 0.69694, 0.70025, 0.70317]
voltOut = [0.64662, 0.63705, .58748, 0.45122, 0.34792, 0.27793, 0.19491]
deltaT = [40, 44, 142, 450, 840, 1260, 2060]

gain = []
for i in range(len(voltIn)):
    gain.append(voltIn[i] - voltOut[i])

plt.figure(figsize = (9, 9))

plt.subplot(2, 1, 1)
plt.plot(freq, gain)
plt.xlabel('Frequency (in Hz)')
plt.ylabel('Voltage (in Volts)')

plt.subplot(2, 1, 2)
plt.plot(freq, deltaT)
plt.xlabel('Frequency (in Hz)')
plt.ylabel('Delta t (in microseconds)')
```
