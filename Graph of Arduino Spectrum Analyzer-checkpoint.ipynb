```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
freq = [50, 100, 175, 250, 343, 400, 500, 800, 1600, 2500]
volt = [.1348, .262, .4251, .5446, .6539, .6987, .7579, .83975, .88205, .897]
phase = [82.192, 74.11, 63.36, 54.363, 45.491, 41.116, 34.95, 23.715, 12.6605, 8.5]

plt.figure(figsize = (9, 9))

plt.subplot(2, 1, 1)
plt.plot(freq, volt)
plt.xlabel('Frequency (in Hz)')
plt.ylabel('Voltage (in Volts)')

plt.subplot(2, 1, 2)
plt.plot(freq, phase)
plt.xlabel('Frequency (in Hz)')
plt.ylabel('Phase (in degrees)')
```
